我认为最美的公式是**欧拉恒等式（Euler's Identity）**：

$$
e^{i\pi} + 1 = 0
$$

这个公式被称为“上帝创造的公式”，它将数学中五个最具代表性的基础常数完美融为一体：
- $e$：自然常数，描述连续增长与衰减的核心常量，是微积分的基石之一
- $i$：虚数单位，拓展了数域的边界，是复分析、信号处理等领域的基础
- $\pi$：圆周率，贯穿几何、周期运动、概率论等众多数学分支
- $1$：自然数的基本单位，乘法运算的单位元
- $0$：加法运算的单位元，代表“空”与“起点”的数学抽象

它是通用欧拉公式 $e^{ix} = \cos x + i\sin x$ 当 $x=\pi$ 时的特殊形式。看似毫不相关的指数、三角、代数、几何领域，在这个等式中达成了完美的统一，极致的简洁背后蕴含着极深的数学内核，是数学和谐之美的最佳代表。

```python

%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 2 * np.pi, 200)
y_sin = np.sin(x)
y_cos = np.cos(x)


plt.figure(figsize=(8, 4), dpi=100)
plt.plot(x, y_sin, label='sin(x)', linewidth=2)
plt.plot(x, y_cos, label='cos(x)', linewidth=2, linestyle='--')


plt.title('NumPy 生成的正弦与余弦曲线', fontsize=14)
plt.xlabel('x (弧度)', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()


plt.show()
```
